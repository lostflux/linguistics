### Logistic Regression

```python
from logistic_regression import LogisticRegression
import pandas as pd

# import display, Markdown
from IPython.display import display, Markdown
```

```python
# implement grid search
def grid_search(learning_rates=None, iterations=None, batch_sizes=None):
  # define the hyperparameters
  learning_rates = learning_rates or [ 0.1, 0.01, 0.001 ]
  iterations = iterations or [ 1, 4, 16, 64 ]
  batch_sizes = batch_sizes or [ 1, 4, 32] # , 64, 128, 256, 512, 1024 ]
  # define the best accuracy
  # best_accuracy = 0
  # define the best hyperparameters
  # best_learning_rate = 0
  # best_iteration = 0
  
  # accumulator = pd.DataFrame(columns=["learning_rate", "iterations", "batch_size", "metrics", "f1_score", "precision", "recall", "accuracy"])
  accumulator = pd.DataFrame(columns=["LR", "ITERATIONS", "BATCH-SIZE", "PRECISION-POS", "RECALL-POS", "F1-POS", "PRECISION-NEG", "RECALL-NEG", "F1-NEG", "CONFUSION-MATRIX"])
  
  # ? for small dataset
  accumulator = pd.DataFrame(columns=["LR", "ITERATIONS", "BATCH-SIZE", "PRECISION-POS", "RECALL-POS", "F1-POS"])
  # loop through the learning rates
  
  # create a logistic regression model
  model = LogisticRegression(verbose=False)
  
  # train_set = "movie_reviews_small/train"
  # test_set = "movie_reviews_small/test"
  
  train_set = "movie_reviews/train"
  test_set = "movie_reviews/dev"
  
  for eta in learning_rates:
    # loop through the iterations
    for iteration in iterations:
      
      for batch_size in batch_sizes:
        
        # print(f"{eta, iteration, batch_size = }")
        # train the model
        # model.train()
        model.train(train_set, batch_size=3, n_epochs=iteration, eta=0.1)
        # calculate the accuracy
        
        # results = model.test(test_set)
        
        results = model.evaluate(model.test(test_set))
        
        metrics = results["pos"]["metrics"]
        
        confusion_matrix = results["pos"]["CM"]
        
        # print(results)
        
        # save to df
        accumulator = pd.concat([ accumulator, pd.DataFrame({
          "LR": eta,
          "ITERATIONS": iteration,
          "BATCH-SIZE": batch_size,
          # "METRICS": metrics,
          "PRECISION-POS": metrics["PRECISION"],
          "RECALL-POS": metrics["RECALL"],
          "F1-POS": metrics["F1"],
          
          # "PRECISION-NEG": metrics[0]["PRECISION"],
          # "RECALL-NEG": metrics[0]["RECALL"],
          # "F1-NEG": metrics[0]["F1"],
          # "CONFUSION-MATRIX": confusion_matrix,
          })])
        
        model.reset()
        
  return accumulator
```

```python
result_df = grid_search()
# sort by F1-POS
result_df = result_df.sort_values(by="F1-POS", ascending=False)
display(result_df)
```
