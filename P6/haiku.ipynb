```python
        # SOMETHING2 -> W1 W6 | W2 W5 | W3 W4 | W4 W3 | W5 W2 | W6 W1

raw_s = f"""
        
        S -> VERSION1
        
        VERSION1 -> NP-5 BREAK S-7 BREAK S-5
        
        NP-5 -> DET-1 ADJ-1 ADJ-2 NOUN-1
        
        S-7 -> VP-3 NPOBJ-4
        
        VP-3 -> NP-2 VERB-1
        NPOBJ-4 -> PREP-1 NP-3
        
        S-5 -> S-1 NPOBJ-4
        
        S-1 -> NOUN-1
        
        
        NP-2 -> DET-1 NOUN-1
        
        NP-3 -> DET-1 ADJ-1 NOUN-1 | DET-1 ADJ-2 NOUN-2
        
        
        
        BREAK -> "|"
        
        DET-1 -> { " | ".join([ f"'{word}'" for word in DET1 ]) }
        ADJ-1 -> { " | ".join([ f"'{word}'" for word in ADJ1 ]) }
        ADJ-2 -> { " | ".join([ f"'{word}'" for word in ADJ2 ]) }
        PREP-1 -> { " | ".join([ f"'{word}'" for word in PREP1 ]) }
        CONJ-1 -> { " | ".join([ f"'{word}'" for word in CONJ1 ]) }
        ADV-1 -> { " | ".join([ f"'{word}'" for word in ADV1 ]) }
        NOUN-1 -> { " | ".join([ f"'{word}'" for word in NOUN1 ]) }
        PNOUN-1 -> { " | ".join([ f"'{word}'" for word in PNOUN1 ]) }
        VERB-1 -> { " | ".join([ f"'{word}'" for word in VERB1 ]) }    
    """
    
print(f"{raw_s}\n\n")

cfgGrammarHaiku = CFG.fromstring(raw_s)
```

```python
counter = 0
with open("haiku.txt", "w") as outfile:
    for sentence in generate(cfgGrammarHaiku):
        if counter % 1_000_00 == 0:
            print(f"{counter:10d}: {' '.join(sentence)}")
            outfile.write(' '.join(sentence) + "\n")
        
        if counter > 500_000_000:
            break
        
        counter += 1

print("\nSo far we have generated " + str(counter) + " sentences.")
```
