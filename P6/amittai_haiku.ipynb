```python
from nltk import CFG
from nltk.parse.generate import generate
import pyphen
```

```python
dic = pyphen.Pyphen(lang='en')
def count_syllables(word):
  sylls = dic.inserted(word)
  return sylls.count("-") + 1

```

```python
W1 = {
    "a"
  , "an"
  , "old"
  , "pond"
  , "frog"
  , "jumps"
  , "the"
  , "pond"
  , "splash"
  , "again"
  , "world"
  , "dew"
  , "of"
  , "and"
  , "every"
  , "struggle"
}

W2 = {
    "silent"
  , "within"
  , "dewdrop"
  , "silence"
  , "into"
}


# three-syllable words
W3 = {
    "beautiful", "dangerous", "mysterious"#, "delicious", "curious", "serious", "wonderful"          # adjectives
  , "daughter", "teacher", "manager"#, "visitor", "stranger", "neighbor"                             # nouns
  , "remember", "discover", "consider"#, "encounter", "wander", "listen", "whisper"        # verbs
}

# four-syllable words
W4 = {
    "important", "impossible", "incredible"#, "unbelievable", "unforgettable"                        # adjectives
  , "ambassador", "celebration", "communication"#, "conversation", "entertainment"                   # nouns
  , "understand", "appreciate", "communicate"#, "concentrate", "celebrate", "entertain"              # verbs
}

raw_s = f"""
    S -> SOMETHING1 BREAK SOMETHING2 BREAK SOMETHING3
    
    SOMETHING1 -> W1 W4 | W2 W3 | W3 W2 | W4 W1
    SOMETHING2 -> W3 W2 W3 | W4 W1 W2
    SOMETHING3 -> W4 W1 | W3 W2 | W1 W4 | W2 W3
    
    BREAK -> "|"
    
    W1 -> { " | ".join([ f"'{word}'" for word in W1 ]) }
    W2 -> { " | ".join([ f"'{word}'" for word in W2 ]) }
    W3 -> { " | ".join([ f"'{word}'" for word in W3 ]) }
    W4 -> { " | ".join([ f"'{word}'" for word in W4 ]) }
    """
    
print(f"{raw_s}\n\n")

cfgGrammarHaiku = CFG.fromstring(raw_s)
```

```python
counter = 0
with open("haiku-0.txt", "w") as outfile:
    for sentence in generate(cfgGrammarHaiku):
        if counter % 100_000 == 0:
            print(f"{counter:10d}: {' '.join(sentence)}")
            outfile.write(' '.join(sentence) + "\n")
        
        if counter > 5_000_000:
            break
        
        counter += 1

print("\nSo far we have generated " + str(counter) + " sentences.")
```

```python
raw_s = f"""
        
        S -> NP-5 BREAK S-7 BREAK S-5
        
        NP-5 -> DET-1 ADJ-1 ADJ-2 NOUN-1
        
        S-7 -> VP-3 NPOBJ-4
        
        VP-3 -> NP-2 VERB-1
        NPOBJ-4 -> PREP-1 NP-3
        
        S-5 -> S-1 NPOBJ-4
        
        S-1 -> NOUN-1
        
        
        NP-2 -> DET-1 NOUN-1
        
        NP-3 -> DET-1 ADJ-1 NOUN-1 | DET-1 NOUN-2
        
        
        
        BREAK -> "|"
        
        DET-1 -> { " | ".join([ f"'{word}'" for word in DET1 ]) }
        ADJ-1 -> { " | ".join([ f"'{word}'" for word in ADJ1 ]) }
        ADJ-2 -> { " | ".join([ f"'{word}'" for word in ADJ2 ]) }
        PREP-1 -> { " | ".join([ f"'{word}'" for word in PREP1 ]) }
        CONJ-1 -> { " | ".join([ f"'{word}'" for word in CONJ1 ]) }
        ADV-1 -> { " | ".join([ f"'{word}'" for word in ADV1 ]) }
        NOUN-1 -> { " | ".join([ f"'{word}'" for word in NOUN1 ]) }
        PNOUN-1 -> { " | ".join([ f"'{word}'" for word in PNOUN1 ]) }
        VERB-1 -> { " | ".join([ f"'{word}'" for word in VERB1 ]) }    
    """
    
print(f"{raw_s}\n\n")

cfgGrammarHaiku = CFG.fromstring(raw_s)
```
